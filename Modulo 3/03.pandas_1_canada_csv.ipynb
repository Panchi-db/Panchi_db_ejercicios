#### Extra: Obtener un Dataframe que te parezca relevante a partir de la aplicación de filtros. ¡Se creativo!

```python
df['Total_Migraciones'] = df[range(1980, 2014)].sum(axis=1)

promedio_continente = df.groupby('Continente')['Total_Migraciones'].mean()
continentes_destacados = promedio_continente[promedio_continente > promedio_continente.mean()].index

df_filtro = df[df['Continente'].isin(continentes_destacados)]
umbral_migraciones = df_filtro['Total_Migraciones'].quantile(0.75)
df_filtro = df_filtro[df_filtro['Total_Migraciones'] > umbral_migraciones]
columnas_relevantes = ['Continente', 'Pais', 'Total_Migraciones'] + list(range(2000, 2011))
df_filtro = df_filtro[columnas_relevantes]
df_filtro['Tendencia_Migratoria'] = df_filtro[2010] - df_filtro[2000]
df_filtro = df_filtro.sort_values(by=['Continente', 'Tendencia_Migratoria'], ascending=[True, False]).reset_index(drop=True)


df_filtro
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Crear el gráfico
plt.figure(figsize=(12, 5))
plt.bar(df_filtro['Pais'], df_filtro['Tendencia_Migratoria'], facecolor= 'blue', alpha=0.7, edgecolor='black', linewidth=1.2, capsize=0.3, ecolor='black')
plt.xticks(rotation=90)
plt.title('Tendencia Migratoria por País')
plt.xlabel('País')
plt.ylabel('Tendencia Migratoria')
plt.show()
```
