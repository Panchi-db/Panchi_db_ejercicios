## Carga de datos
Cargar multiples CSVs en un mismo DF de Pandas, concatenando los datos de cada archivo.

```python
#listar los archivos del csv del directorio spotify usando os.
```

```python
file_names = os.listdir ('../Data/spotify')
file_names
```

```python
#bucle for para cargar uno a uno en pandas con un concat para unirlos en un solo DF
df= pd.DataFrame()  #creamos un dataframe vacio donde se iran concatenando los archivos


for file in file_names:
        df_file = pd.read_csv(f'../Data/spotify/' + file)
        print(file, df_file.shape)
        df= pd.concat([df, df_file], ignore_index=True)  
print(f" df final :{df.shape}")     
        
```
