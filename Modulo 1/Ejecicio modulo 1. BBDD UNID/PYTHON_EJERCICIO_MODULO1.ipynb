```python
#Esta parte pasará al fichero alumnos.py
#Creacion de CLASE
from datetime import datetime

class Alumnos:
    
    def __init__ (self, nombre, apellido, telefono, 
                  email, estudiante, not_med, f_cum):
        
        self.nombre= nombre
        self.apellido= apellido
        self.telefono= telefono
        self.email= email
        self.estudiante= estudiante
        self.not_med= not_med
        self.f_cum= datetime.strptime (f_cum, "%d/%m/%Y")
        print("Alta correcta Alumn@")
        
    def calcular_edad (self):
        hoy = datetime.now()
        edad= hoy.year - self.f_cum.year
        if (hoy.month, hoy.day) < (self.f_cum.month, self.f_cum.day):
            edad-=1
        return edad
    def actualizar_alumno (self, dato, nuevo_dato):
        datos_actualizables= ["nombre", "apellido", "telefono", "email"]
        if dato in datos_actualizables:
            setattr(self, dato, nuevo_dato)
            print (f" El {dato} se ha actualizado a : {nuevo_dato}")
        else:
            print ("El dato no es actualizable")
    def __str__(self):
        return f"""Alumn@.-\n Nombre: {self.nombre} \n Apellido: {self.apellido} \n Teléfono: {self.telefono}\n Email: {self.email}\n Edad: {self.calcular_edad()} \n Estudiante: {self.estudiante} \n Nota media:{self.not_med} \n Cumpleños: {self.f_cum.strftime('%d/%m/%y')}\n  """
        
    def __repr__(self):
        return f" Alumn@s: (Nombre: {self.nombre}, Apellido: {self.apellido}, Teléfono: {self.telefono}, Email: {self.email}, Edad: {self.calcular_edad()}, Estudiante: {self.estudiante}, Nota media:{self.not_med} Cumpleaños: {self.f_cum.strftime('%d/%m/%y')})" 
    
       
```

```python
#Str menu va junto con el menu corto con las funciones en def, en el fichero app.py
menu =  """WEB de ALUMN@S de la UNID. \n OPCIONES: \n 1.- Listado completo de alumn@s \n 2.- Listado de alumn@s por edad \n 3.- Estado Administrativo \n 4.- Buscar almun@ por email \n 5.- Alta nuev@ alumn@ \n 6.- Actualizar datos alumn@ \n 7.- Baja alumn@s \n 8.- Eliminar alumnado \n 9.- Salir de la WEB."""
print (menu)
```

```python
def f_cum(alumno):
    return alumno.f_cum
```

```python
#Al lio

while True:

    opcion = int(input(menu))
    match opcion:
        case 1:
            print ("1.- Listado completo de alumn@s ")
            for alumno in alumnos:
                print (f" {alumno}")     
            if not alumnos:
                print ("No existen alumn@s.")
            print ("\n")
            break  #Sí funciona   
            
        case 2:
            print ("2.- Listado de alumn@s por edad")
            alumnos_por_edad= sorted (alumnos, key= lambda alumno : alumno.f_cum)
            for alumno in alumnos_por_edad:
                print (alumno)
            print("\n")
            break #si lo pongo , se para en el primer registro y no da los otros 3
        case 3:
            print ("3.- Estado Administrativo")
            for alumno in alumnos:
                if alumno.estudiante:
                    print (f"{alumno.nombre} {alumno.apellido}: Alta")
                else:
                    print (f"{alumno.nombre} {alumno.apellido}: Baja")
            print("\n")
            break # Sí funciona
       
        case 4:
            print ("4.- Buscar almun@ por email")
            email= input("Introduce un email:")
            found= False
            for alumno in alumnos:
                if email.lower() == alumno.email.lower():
                    print(alumno)
                    found= True
                    print ("\n")
                    break #No funciona
                if not found:
                    print ("El email no existe en la BBDD. Escríbelo bien:")
            print("\n")
            break # Sí funciona
                                  
        case 5:
            print ("5.- Alta nuev@ alumn@")
            nombre= input("Introduce el nombre:")
            apellido= input("Introduce el apellido:")
            telefono= input("Introduce el teléfono: \n XXXXXXXXX")
            email= input("Introduce el email: ")
            not_med= float(input("Introduce la nota media:"))
            estudiante= bool(input("¿Está matriculad@? \n Opciones: True/False"))
            f_cum= input ("Introduce la fecha de nacimiento: \n Formto dd/mm/aaa")
            nuevo_alumno= Alumnos (nombre, apellido, telefono , email, estudiante, not_med, f_cum)
            alumnos.append(nuevo_alumno)
            print(nuevo_alumno)
            break # Sí funciona
            
        case 6:
            print ("6.- Actualizar datos alumn@")
            if not alumnos:
                print ("No hay alumn@s dados de alta")
                break # Sí funciona
            else:
                email= input("Introduce el email del alumn@ a actualizar")
                found= False
                for alumno in alumnos:
                    if alumno.email.lower()== email.lower():
                        found= True
                        print(f" Alumno encontrado: \n {alumno}")
                        dato= input("Para actualizar Nombre, escribe la plabra 'Nombre', no el nombre concreto. \n Igual con el resto de datos: 'Apellido', 'Teléfono' (sin acento), 'Email' que quieras actualizar: ").strip().lower()
                        if dato in ["nombre", "apellido", "telefono", "email"]:
                            nuevo_dato= input(f"Nuevo dato: {dato}")
                            alumno.actualizar_alumno(dato, nuevo_dato)
                            print ("Dato actualizado correctamente:")
                            print(alumno)
                            break # No funciona
                        else:
                            print(f" El dato {dato} no es modificable")
                            break # No funciona 
          

        case 7:
            print ("7.- Baja alumn@s ")
            if not alumnos:
                print("No existen alumn@s")
                break # Sí funciona
            else:
                baja= input("Introduce el nombre del alumno: ").strip().lower()
                found= False
                for alumno in alumnos:
                    if alumno.nombre.lower() == baja.lower():
                        alumnos.remove(alumno)
                        print (f" Alumn@ {alumno.nombre} {alumno.apellido} ha sido dado de baja")
                        found= True
                        break # No funciona 
                if not found:
                    print (f"No se ha encontrado al alumn@ {baja}")
                    break  # Sí funciona   
        case 8:
            print ("8.- Eliminar alumnado")
            alumnos.clear()
            print ("Todos los alumn@s han sido dad@s de baja correctamente.\nPara nuevo alta selecciona opción 4 y sigue las intrucciones.")
            break # Sí funciona
        
        case 9:
            print ("9.- Has salido de la WEB. \n    ¡Hasta pronto!.")
            break # Sí funciona
             
        case _:
            print ("    ¡ERROR FATAL!\n    Elige otra opción.")
            break  # Sí funciona       
```
