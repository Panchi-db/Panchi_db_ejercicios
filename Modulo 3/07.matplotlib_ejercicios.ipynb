### Ejercicio 10:
- Muestra en 2 graficas de sectores con _**plt.subplots()**_ la siguiente información:
    1. En la primera gráfica mostrar el número de pasajeros que murieron de cada clase.
    2. En la segunda gráfica mostrar el número de pasajeros que sobrevivieron de cada clase.

<div align="center">
   
</div>

```python
labels = sorted([f"clase {i}" for i in df_tit['Pclass'].unique()])
plt.subplot(1, 2, 1)
values = df_tit[df_tit['Survived'] == 0].value_counts('Pclass', sort=False, normalize=True)
plt.pie(x=values, 
        labels=labels, 
        colors=['green', 'red', 'cyan'], 
        explode = (0.1, 0.1, 0.1),
        autopct = "%1.2f%%")
plt.subplot(1, 2, 2)
values = df_tit[df_tit['Survived'] == 1].value_counts('Pclass', sort=False, normalize=True)
plt.pie(x=values, 
        labels=labels, 
        colors=['green', 'red', 'cyan'], 
        explode = (0.1, 0.1, 0.1),
        autopct = "%1.2f%%")
plt.title("Not survived vs. Survived %");
```
